```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### DATASET NETFLIX STOCK PRICE PREDICTION

```python
netflix_ds = pd.read_csv("DataSet\\Netflix\\NFLX.csv")
```

```python
netflix_ds.info()
```

```python
netflix_ds.head()
```

```python
netflix_ds['Date'] = pd.to_datetime(netflix_ds['Date'])
```

```python
netflix_ds['Date'].describe()
```

```python
netflix_ds['Date'].info()
```

```python
netflix_ds.set_index('Date', inplace=True)
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout = True, figsize=(15, 8))
netflix_ds['Close'].plot(ax=axs[0], xlabel="Date", ylabel="Price", title="Close Prices")
netflix_ds['Open'].plot(ax=axs[1], xlabel="Date", ylabel="Price", title="Open Prices", color="teal")
```
